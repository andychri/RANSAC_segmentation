```python
import numpy as np
import open3d as o3d
from scipy.optimize import minimize
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

def cone_distance(params, points):
    vertex = params[:3]
    axis = params[3:6]
    angle = params[6]

    axis = axis / np.linalg.norm(axis)
    vectors = points - vertex
    distances = np.linalg.norm(vectors, axis=1)
    cos_angles = np.dot(vectors, axis) / distances
    cone_distances = np.abs(cos_angles - np.cos(angle))

    return np.sum(cone_distances ** 2)

def fit_cone_to_points(points):
    initial_vertex = np.mean(points, axis=0)
    initial_axis = np.array([0, 0, 1])
    initial_angle = np.pi / 6
    initial_params = np.hstack((initial_vertex, initial_axis, initial_angle))

    bounds = [(-np.inf, np.inf), (-np.inf, np.inf), (-np.inf, np.inf),
              (-1, 1), (-1, 1), (-1, 1),
              (0, np.pi / 2)]

    result = minimize(cone_distance, initial_params, args=(points,), bounds=bounds, method='L-BFGS-B')

    if result.success:
        fitted_params = result.x
        vertex = fitted_params[:3]
        axis = fitted_params[3:6] / np.linalg.norm(fitted_params[3:6])
        angle = fitted_params[6]
        return vertex, axis, angle
    else:
        raise RuntimeError("Cone fitting optimization failed")

def calculate_inliers(points, vertex, axis, angle, threshold):
    vectors = points - vertex
    distances = np.linalg.norm(vectors, axis=1)
    cos_angles = np.dot(vectors, axis) / distances
    cone_distances = np.abs(cos_angles - np.cos(angle))

    return cone_distances < threshold

def ransac_cone(points, num_iterations=1000, threshold=0.05):
    best_inliers = []
    best_cone = None

    for _ in range(num_iterations):
        sample_indices = np.random.choice(points.shape[0], 3, replace=False)
        sample_points = points[sample_indices]

        try:
            vertex, axis, angle = fit_cone_to_points(sample_points)
            inliers = calculate_inliers(points, vertex, axis, angle, threshold)

            if inliers.sum() > len(best_inliers):
                best_inliers = inliers
                best_cone = (vertex, axis, angle)
        except RuntimeError:
            continue

    return best_cone, best_inliers

def plot_cone(ax, vertex, axis, angle, height, resolution=50):
    theta = np.linspace(0, 2 * np.pi, resolution)
    r = height * np.tan(angle)
    x = r * np.cos(theta)
    y = r * np.sin(theta)
    z = height * np.ones_like(x)

    cone_points = np.vstack((x, y, z)).T

    rotation_matrix, _ = np.linalg.qr(np.column_stack((axis, np.random.randn(3), np.random.randn(3))))
    rotated_cone_points = cone_points.dot(rotation_matrix.T) + vertex

    poly3d = [[tuple(rotated_cone_points[j]) for j in range(len(rotated_cone_points))]]
    ax.add_collection3d(Poly3DCollection(poly3d, color='cyan', alpha=0.3, edgecolors='r'))

    line_points = np.array([vertex, vertex + height * axis])
    ax.plot(line_points[:, 0], line_points[:, 1], line_points[:, 2], color='r')

def plot_point_cloud_and_cone(points, cone_params, inliers):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(points[:, 0], points[:, 1], points[:, 2], c='b', marker='o', alpha=0.5)
    ax.scatter(points[inliers, 0], points[inliers, 1], points[inliers, 2], c='r', marker='o', alpha=0.5)

    vertex, axis, angle = cone_params
    height = 10
    plot_cone(ax, vertex, axis, angle, height)

    plt.show()

if __name__ == "__main__":
    # Load a point cloud from a PCD file
    pcd = o3d.io.read_point_cloud("filtered.pcd")
    point_cloud = np.asarray(pcd.points)

    best_cone, inliers = ransac_cone(point_cloud)

    print("Best Cone Parameters:")
    print("Vertex:", best_cone[0])
    print("Axis:", best_cone[1])
    print("Angle:", best_cone[2])
    print("Number of inliers:", np.sum(inliers))

    plot_point_cloud_and_cone(point_cloud, best_cone, inliers)
```

```python
import numpy as np
import open3d as o3d
from scipy.optimize import minimize

def cone_distance(params, points):
    vertex = params[:3]
    axis = params[3:6]
    angle = params[6]

    axis = axis / np.linalg.norm(axis)
    vectors = points - vertex
    distances = np.linalg.norm(vectors, axis=1)
    cos_angles = np.dot(vectors, axis) / distances
    cone_distances = np.abs(cos_angles - np.cos(angle))

    return np.sum(cone_distances ** 2)

def fit_cone_to_points(points):
    initial_vertex = np.mean(points, axis=0)
    initial_axis = np.array([0, 0, 1])
    initial_angle = np.pi / 6
    initial_params = np.hstack((initial_vertex, initial_axis, initial_angle))

    bounds = [(-np.inf, np.inf), (-np.inf, np.inf), (-np.inf, np.inf),
              (-1, 1), (-1, 1), (-1, 1),
              (0, np.pi / 2)]

    result = minimize(cone_distance, initial_params, args=(points,), bounds=bounds, method='L-BFGS-B')

    if result.success:
        fitted_params = result.x
        vertex = fitted_params[:3]
        axis = fitted_params[3:6] / np.linalg.norm(fitted_params[3:6])
        angle = fitted_params[6]
        return vertex, axis, angle
    else:
        raise RuntimeError("Cone fitting optimization failed")

def calculate_inliers(points, vertex, axis, angle, threshold):
    vectors = points - vertex
    distances = np.linalg.norm(vectors, axis=1)
    cos_angles = np.dot(vectors, axis) / distances
    cone_distances = np.abs(cos_angles - np.cos(angle))

    return cone_distances < threshold

def ransac_cone(points, num_iterations=1000, threshold=0.05):
    best_inliers = []
    best_cone = None

    for _ in range(num_iterations):
        sample_indices = np.random.choice(points.shape[0], 3, replace=False)
        sample_points = points[sample_indices]

        try:
            vertex, axis, angle = fit_cone_to_points(sample_points)
            inliers = calculate_inliers(points, vertex, axis, angle, threshold)

            if inliers.sum() > len(best_inliers):
                best_inliers = inliers
                best_cone = (vertex, axis, angle)
        except RuntimeError:
            continue

    return best_cone, best_inliers

def create_cone_mesh(vertex, axis, angle, height, resolution=50):
    if height <= 0:
        raise ValueError("Height must be positive.")
    
    # Calculate the radius based on the cone angle and height
    radius = height * np.tan(angle)
    if radius <= 0:
        raise ValueError("Cone radius must be positive.")

    mesh = o3d.geometry.TriangleMesh.create_cone(radius=radius, height=height, resolution=resolution)
    rotation_matrix = np.linalg.qr(np.column_stack((axis, np.random.randn(3), np.random.randn(3))))[0]
    mesh.rotate(rotation_matrix, center=(0, 0, 0))
    mesh.translate(vertex)
    return mesh

def visualize_cone_fitting(point_cloud, inliers, best_cone):
    # Create a point cloud for inliers
    inlier_cloud = o3d.geometry.PointCloud()
    inlier_cloud.points = o3d.utility.Vector3dVector(point_cloud[inliers])

    # Create a point cloud for outliers
    outlier_cloud = o3d.geometry.PointCloud()
    outlier_cloud.points = o3d.utility.Vector3dVector(point_cloud[~inliers])
    outlier_cloud.paint_uniform_color([1, 0, 0])  # Red color for outliers

    # Create a cone mesh
    vertex, axis, angle = best_cone
    height = 10  # Adjust this value based on your point cloud
    cone_mesh = create_cone_mesh(vertex, axis, angle, height)

    # Visualize the point cloud and the cone
    o3d.visualization.draw_geometries([inlier_cloud, outlier_cloud, cone_mesh])

if __name__ == "__main__":
    # Load a point cloud from a PCD file
    pcd = o3d.io.read_point_cloud("filtered.pcd")
    point_cloud = np.asarray(pcd.points)

    best_cone, inliers = ransac_cone(point_cloud)

    if best_cone is not None:
        print("Best Cone Parameters:")
        print("Vertex:", best_cone[0])
        print("Axis:", best_cone[1])
        print("Angle:", best_cone[2])
        print("Number of inliers:", np.sum(inliers))

        visualize_cone_fitting(point_cloud, inliers, best_cone)
    else:
        print("No cone found.")
```
